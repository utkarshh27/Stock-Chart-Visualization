```python
import pandas as pd
import matplotlib.pyplot as plt


stock_data = pd.read_csv(r'C:\Users\UTKARSH SINGH\Jupiter Porjects\stock_data.csv')
stock_data = stock_data.dropna()
#stock_data.head()
stock_data['date'] = pd.to_datetime(stock_data['date'], format='%d-%m-%Y')
stock_data.set_index('date', inplace=True)


stock_data['10-day MA'] = stock_data['price'].rolling(window=10).mean()
plt.plot(stock_data['price'])
plt.plot(stock_data['10-day MA'])
plt.title('Stock Prices with 10-day Moving Average')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.legend(['Price', '10-day MA'])
plt.show()
```
